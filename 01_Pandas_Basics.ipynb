## Indexing and Slicing with reindex()
- reindex메소드는 datetime 날짜시간 인덱스가 있는 시계열 데이터를 작업할 때 주로 씀
- loc()와 다르게 존재하지 않는 행이날 열도 전달 가능
- 한개만 검색할때도 리스트 사용!

```python
summer = pd.read_csv("summer.csv")
```

```python
summer
```

```python
summer.loc[[0, 5, 30000], ["Athlete", "Medal"]]
```

```python
# summer.loc[[0, 5, 30000, 40000], ["Athlete", "Medal"]] # 40000행은 존재 않함
```

```python
summer.reindex(index = [0, 5, 30000, 40000], columns =  ["Athlete", "Medal", "Age"])
# 존재하지 않는 행과 열 추가
# 고유한 값을 갖고있는 행과 열을 갖을때만 reindex() 사용 가능
```

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete")
```

```python
summer
```

```python
summer.reindex(columns = ["Medal", "Age"]) # 존재하는 컬럼만 넣으면 평범하게 검색됨
```

```python
#summer.reindex(index = ["PHELPS, Michael"], columns = ["Medal", "Age"])
# 존재하지 않는 행과 열을 검색할때
# 중복된 PHELPS, Michael가 있으므로 에러
```
